## Dummyficar respuestas múltiples

```python
columns = [
    'cambio_climatico',
    'crisis_climatica',
    'aparatos_electronicos',
    'priorizar_inversiones',
    'calidad_aire',
    'valoras_mas',
    'sensibilidad_climatica',
    'afectar_cambio_climatico',
    'lider_lucha_crisis',
    'afectadas_cambio_climatico',
    'cosas_casa_transporte',
    'futuro_inmediato',
]
```

```python
#Separa las respuestas por pipelines a las preguntas que tienen varias opciones de respuesta para luego dummificar como varias respuestas individuales
def nuevas_columnas(df, column_name):
    df_salida = df[column_name].str.replace("\s+(?=\|)", "", regex=True).str.get_dummies()
    df_salida.columns = [f"{column_name}_{col}" for col in df_salida.columns]
    df_salida = clean_columns(df_salida)
    df_salida = pd.concat(
    (df.drop([column_name], axis=1), df_salida),
    axis=1,)
    return df_salida
```

```python
#Aplica la función anterior por cada columna definida en la lista de columnas
for i in range(len(columns)):
  df_renamed = nuevas_columnas(df_renamed, columns[i])
```

## Lemmatization
La lematización es un proceso lingüístico que consiste en, dada una forma flexionada, hallar el lema correspondiente. El lema es la forma que por convenio se acepta como representante de todas las formas flexionadas de una misma palabra.

Por ejemplo, las palabras niños, niña y niñito tienen todas el mismo lema: niño.

```python
import spacy.cli
spacy.cli.download("es_core_news_sm")
```

```python
import spacy
nlp = spacy.load('es_core_news_sm')

def lemmatizer(text):
  doc = nlp(text)
  return ' '.join([word.lemma_ for word in doc])
```

```python
text = "personas ideas cosas"
print(lemmatizer(text))
```

```python
df_renamed['transicion_ecologica_sin_stopwords_lemma'] = df_renamed['transicion_ecologica_sin_stopwords'].apply(lambda x: lemmatizer(x))
df_renamed['medida_concreta_sin_stopwords_lemma'] = df_renamed['medida_concreta_sin_stopwords'].apply(lambda x: lemmatizer(x))
df_renamed['accion_transicion_climatica_sin_stopwords_lemma'] = df_renamed['accion_transicion_climatica_sin_stopwords'].apply(lambda x: lemmatizer(x))
df_renamed['ley_congreso_crisis_climatica_sin_stopwords_lemma'] = df_renamed['ley_congreso_crisis_climatica_sin_stopwords'].apply(lambda x: lemmatizer(x))
df_renamed['haria_falta_futuro_sin_stopwords_lemma'] = df_renamed['haria_falta_futuro_sin_stopwords'].apply(lambda x: lemmatizer(x))
```
